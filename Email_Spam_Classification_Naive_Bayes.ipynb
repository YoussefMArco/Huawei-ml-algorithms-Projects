```python
y=df["Prediction"]
df.drop("Prediction",axis=1)
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB, GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# Initialize models
models = {
    'MultinomialNB': MultinomialNB(),
    'BernoulliNB': BernoulliNB(),
    'GaussianNB': GaussianNB(),  # Fixed GaussianNB initialization
    'SVM': SVC(kernel='linear')  # Support Vector Machine
}

# Iterate through models and fit them
for model_name, model in models.items():
    print(f"Training {model_name}...")
    model.fit(X_train, y_train)  # Train the model
    accuracy = model.score(X_test, y_test)  # Evaluate accuracy on test data
    print(f"{model_name} Accuracy: {accuracy:.2f}\n")
    
    # Make predictions on the test set
    y_pred = model.predict(X_test)
    
    # Print the classification report
    print(f"Classification Report for {model_name}:\n")
    print(classification_report(y_test, y_pred))
```
